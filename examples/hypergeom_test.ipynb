```python
significantGenes = set(de['ttest']['gene'])
```

```python
def hypergeom_test(pw_theo, testSets, print_df=True):
    data = np.zeros((len(testSets),6))
    populationSize = len(pw_theo.protein2mass)
    numSuccInPopulation = len(significantGenes)
    counter = 0
    for setElem in testSets:

        setData = testSets[setElem]
        sampleSize = len(setData)
        successIntersection = significantGenes.intersection(setData)
        drawnSuccesses = len(successIntersection)
        
        data[counter][0] = drawnSuccesses - 1
        data[counter][1] = populationSize
        data[counter][2] = numSuccInPopulation
        data[counter][3] = sampleSize
        pval = hypergeom.sf(drawnSuccesses - 1, populationSize, numSuccInPopulation, sampleSize)
        data[counter][4] = pval
        fractionOfHitSamples = drawnSuccesses / sampleSize if sampleSize > 0 else 0
        data[counter][5] = fractionOfHitSamples
        counter += 1
    if print_df:
        return pd.DataFrame(data=data, index=pw_theo.category2id.keys(), columns=["drawnSuccesses - 1", "populationSize", "numSuccInPopulation", "sampleSize", "pval", "fractionOfHitSamples"])
```

```python
hypergeom_test(pw_theo, pw_theo.category2id)
```
